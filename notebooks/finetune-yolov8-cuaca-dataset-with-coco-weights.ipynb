```python
import numpy as np
import pandas as pd

# Retrieve class names
class_names = metrics.names
num_classes = len(class_names)

# Get the per-class metrics by calling metrics.summary() and convert to DataFrame
per_class_summary = metrics.summary()
per_class_df = pd.DataFrame(per_class_summary)

# Initialize lists to store per-class metrics
per_class_data = []

# Get the confusion matrix
cm_matrix = metrics.confusion_matrix.matrix

# Calculate per-class metrics
for i, class_name in class_names.items():
    # True Positives: diagonal elements
    TP_i = cm_matrix[i, i]

    FP_i = np.sum(cm_matrix[:, i]) - TP_i

    FN_i = np.sum(cm_matrix[i, :]) - TP_i

    TN_i = np.sum(cm_matrix) - (TP_i + FP_i + FN_i)

    # Access precision, recall, mAP50, and mAP50-95 from per_class_df
    # Ensure class 'i' exists in the per_class_df index before accessing
    if i in per_class_df.index:
        precision_i = per_class_df.loc[i, 'Box-P']
        recall_i = per_class_df.loc[i, 'Box-R']
        map50_i = per_class_df.loc[i, 'mAP50']
        map50_95_i = per_class_df.loc[i, 'mAP50-95']
    else:
        # Handle cases where a class might not appear in per_class_df (e.g., no instances)
        precision_i = 0.0
        recall_i = 0.0
        map50_i = 0.0
        map50_95_i = 0.0

    # Calculate F1-score
    f1_score_i = 2 * (precision_i * recall_i) / (precision_i + recall_i) if (precision_i + recall_i) > 0 else 0.0
    
    # Sensitivity is equivalent to Recall
    sensitivity_i = recall_i

    specificity_i = TN_i / (TN_i + FP_i) if (TN_i + FP_i) > 0 else 0.0
    accuracy_i = (TP_i + TN_i) / (TP_i + FP_i + FN_i + TN_i) if (TP_i + FP_i + FN_i + TN_i) > 0 else 0.0

    per_class_data.append({
        'Class': class_name,
        'TP': int(TP_i),
        'FP': int(FP_i),
        'FN': int(FN_i),
        'TN': int(TN_i),
        'Precision': precision_i,
        'Recall': recall_i,
        'F1-Score': f1_score_i,
        'Accuracy': accuracy_i,
        'Sensitivity': sensitivity_i,
        'Specificity': specificity_i,
        'mAP50': map50_i,
        'mAP50-95': map50_95_i
    })

# Overall metrics
overall_precision = metrics.box.mp
overall_recall = metrics.box.mr
overall_f1_score = 2 * (overall_precision * overall_recall) / (overall_precision + overall_recall) if (overall_precision + overall_recall) > 0 else 0.0
overall_mAP50 = metrics.box.map50
overall_mAP50_95 = metrics.box.map


overall_TP = np.sum([cm_matrix[i, i] for i in range(num_classes)])
overall_FP = np.sum([np.sum(cm_matrix[:, i]) - cm_matrix[i, i] for i in range(num_classes)])
overall_FN = np.sum([np.sum(cm_matrix[i, :]) - cm_matrix[i, i] for i in range(num_classes)])

overall_TN = np.sum(cm_matrix) - (overall_TP + overall_FP + overall_FN)

overall_accuracy = (overall_TP + overall_TN) / (overall_TP + overall_FP + overall_FN + overall_TN) if (overall_TP + overall_FP + overall_FN + overall_TN) > 0 else 0.0
overall_sensitivity = overall_recall # By definition, sensitivity is recall
overall_specificity = overall_TN / (overall_TN + overall_FP) if (overall_TN + overall_FP) > 0 else 0.0


# Create DataFrame for per-class report
df_per_class = pd.DataFrame(per_class_data).round(3)

# Create DataFrame for overall report
overall_data = {
    'Metric': ['Precision', 'Recall', 'F1-Score', 'Accuracy', 'Sensitivity', 'Specificity', 'mAP50', 'mAP50-95'],
    'Value': [
        overall_precision,
        overall_recall,
        overall_f1_score,
        overall_accuracy,
        overall_sensitivity,
        overall_specificity,
        overall_mAP50,
        overall_mAP50_95
    ]
}
df_overall = pd.DataFrame(overall_data).round(3)

print("### Comprehensive Classification Report ###")
print("\n--- Per-Class Metrics ---")
print(df_per_class.to_string(index=False))

print("\n--- Overall Metrics ---")
print(df_overall.to_string(index=False))

print("\nReport generation complete.")
```
